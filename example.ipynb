Как определить «успешный» проект? Наша команда выделяет три метрики «усмешности» проекта:

<ol>
    <li>Финансовый успех — показатель ROI</li>
    <li>Творческий успех — взвешенный рейтинг фильма</li>
    <li>Вовлеченность — число просмотренных минут зрителями, оценившими проект.</li>
</ol>

Высокорейтинговые фильмы повышают статус студии, привлекают известных режиссёров и актеров, помогают получать инвестиции и определяют успех будущих проектов. Стриминг-сервисам могут быть полезны проекты с высокой вовлеченностью, поскольку для них важны просмотры, обсуждаемость и удержание аудитории.
